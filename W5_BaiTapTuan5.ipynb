Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int: 

```python
def Max_Of_Two(a,b):
    if a>b:
        return a
    return b

assert Max_Of_Two(9, 10) == 10
assert Max_Of_Two(0, -2) == 0
assert Max_Of_Two(-110, -2) == -2
print("Pass all test")
```

### W5A3 
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
import math
def la_snt(n):
    if n < 2:
        return False
    else:
        for i in range (2, math.isqrt(n)+1):
            if n % i == 0:
                return False
        return True
n = int(input())
print(la_snt(n))
```

### W5A4 
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_hoan_hao(n):
    tong_uoc = 0
    for i in range(1,n):
        if n % i == 0:
            tong_uoc += i
    if tong_uoc == n:
        return True
    else:
        return False
n = int(input())
print(so_hoan_hao(n))
```

### W5A6 
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(n):
    giai_thua = 1
    if n > 0:
        while n > 0:
            giai_thua *= n
            n -= 1
        return giai_thua
    return None

assert(giai_thua(1)) == 1
assert(giai_thua(3)) == 6
assert(giai_thua(5)) == 120
assert(giai_thua(-10)) == None

n = int(input())
giai_thua(n)
```

### W5A7
Viết hàm giả lập một máy tính bỏ túi.
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat \)  
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:  
Phép trừ \( - \)  
Phép cộng \( + \)  
Phép chia \( / \)  
Phép nhân \( * \)  
Kết quả làm tròn đến chữ số thập phân thứ hai.  
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).  
Đầu ra: Kết quả phép toán

```python
def compute(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b != 0:
            return a / b
        
assert(compute(5, 10, "-")) == -5.00
assert(compute(27, 4, "/")) == 6.75
assert(compute(13, 11, "+")) == 24.00

a, b = map(float, input().split())
op = input()
print(f"{compute(a, b, op):.2f}")
```

### W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng.  
Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.  
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này

```python
def bits(n):
    bits = 0
    while n > 0:
        bits = bits * 10 + n % 2
        n //= 2
    return str(bits)[::-1]

def ten_to_bin(n):
    bi = ""
    while n > 0:
        bi = chr(n%2 + ord('0')) + bi
        n//=2
    return bi

assert(ten_to_bin(1)) == "1"
assert(ten_to_bin(4)) == "100"
assert(ten_to_bin(6)) == "110"
assert(ten_to_bin(13)) == "1101"
assert(ten_to_bin(10)) == "1010"
assert(ten_to_bin(100)) == "1100100"
```

```python
def ten_to_bin(n):
    bi = ""
    while n > 0:
        bi = chr(n%2 + ord('0')) + bi
        n//=2
    return bi

m, n = map(int, input().split())
m_bin = ten_to_bin(m)
n_bin = ten_to_bin(n)

while len(m_bin) < len(n_bin):
    m_bin = "0" + m_bin
while len(n_bin) > len(n_bin):
    n_bin = "0" + n_bin

print(m_bin, n_bin)

cnt = 0
for i in range(len(m_bin)):
    if m_bin[i] != n_bin[i]:
        cnt += 1
print(cnt)
```

### W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó.  
Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'.  
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\).   
Bạn có thể giả sử hai từ đầu vào có cùng độ dài.

```python
def dang_cau(a, b):
    ab = {}
    ba = {}
    for char_a, char_b in zip(a, b):
        if char_a in ab:
            if ab[char_a] != char_b:
                return False
        else:
            ab[char_a] = char_b

        if char_b in ba:
            if ba[char_b] != char_a:
                return False
        else:
            ba[char_b] = char_a
    return True
a, b = input().split()
print(dang_cau(a, b))
```

### W5A15 
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triagle(a, b, c):
    if a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or c**2 + a**2 == b**2:
        return True
    else:
        return False
a, b, c = map(int, input().split())
print(is_right_triagle(a, b, c))
```
