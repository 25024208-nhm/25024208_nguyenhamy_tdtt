```python
import math
a,b,c = map(int, input("Nhap vao 3 so: ").split())
if a+b>c and b+c>a and c+a>b:
    p = (a+b+c)/2
    s = math.sqrt(p*(p-a)*(p-b)*(p-c))
    print("Dien tich tam giac la: ", "{:.1f}".format(s))
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
n = int(input("Nhap mot so nguyen (0-9): "))
match n:
    case 0:
        print("Khong")
    case 1:
        print("Mot")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bon")
    case 5:
        print("Nam")
    case 6:
        print("Sau")
    case 7:
        print("Bay")
    case 8:
        print("Tam")
    case 9:
        print("Chin")
```

```python
n = int(input())
def digit(n):
    match n:
        case 0: return "Khong"
        case 1: return "Mot"
        case 2: return "Hai"
        case 3: return "Ba"
        case 4: return "Bon"
        case 5: return "Nam"
        case 6: return "Sau"
        case 7: return "Bay"
        case 8: return "Tam"
        case 9: return "Chin"
        case _: return "Khong tim thay"
print(digit(n))
```
