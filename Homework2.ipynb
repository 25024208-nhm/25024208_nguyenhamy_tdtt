### Bài 3: 
Tính tam giác: Nhâp vào ba số a,b,c bất kì.  
Kiểm tra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường.  
Tính diện tích, chu vi của tam giác và in ra mànhình

```python
a, b, c = map(int, input().split())
if a+b>c and a+c>b and b+c>a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print("Tam giac can")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")
p = (a+b+c)/2
print((p*(p-a)*(p-b)*(p-c))**0.5)
print(a+b+c)
```

### Bài 5: 
Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(int, input().split())
delta = b**2 - 4*a*c
if delta < 0:
    print("Vo nghiem")
elif delta == 0:
    print("Nghiem kep: ", -b/(2*a))
else:
    print("Nghiem thu nhat: ", (-b + delta**0.5)/(2*a))
    print("Nghiem thu hai: ", (-b - delta**0.5)/(2*a))
```

### Bài 6: 
Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bànphím).

```python
a, b, c, d = map(int, input().split())
print(max(a, b, c, d))
```

```python
a, b, c, d = map(int, input().split())
max = a
if b>max:
    max = b
if c>max:
    max = c
if d>max:
    max = d
print(max)
```

```python
a, b, c, d = map(int, input().split())
for i in [a, b, c, d]:
    if i > max:
        max = i
print(max)
```
