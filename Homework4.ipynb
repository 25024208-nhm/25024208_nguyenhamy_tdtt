### W4A18 
Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int, input().split())
for i in range(1, min(a,b)+1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

### W4A19 [EvenNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để in ra các số dương chẵn nhỏ hơn hoặc bằng n (n <= 1000) 


```python
n = int(input())
for i in range(1, n+1):
    if i % 2 == 0:
        print(i, end=" ")

while n>0:
    if n % 2 == 0:
        print(n, end=" ")
    n -= 1
```

### W4A20 
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
while n % 2 == 0:
    n //= 2
if n == 1:
    print("True")
else:
    print("False")
```

### W4A21 
Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
tong = 0
while n != 0:
    tong += n%10
    n //= 10
print(tong)
```

### W4A22 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
while n != 0:
    so = n % 10
    n //= 10
    if so % 2 == 0:
        chan += 1
    else:
        le += 1
print(chan, "chan", le, "le")
```

### W4A24

1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = int(input())
n = 0
S = 0
while S <= A:
    n += 1
    S += 1/n
print(n)
```

### W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
min = 0
max = 0
while True:
    n = int(input())
    if n != -1:
        if n > max:
            max = n
        if n < min:
            min = n
        continue
    break
print(max, min)
```

### W4A26 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó  
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
f1, f2 = 1, 1
while f1+f2 <= A:
    f_next = f1 + f2
    f1 = f2
    f2 = f_next
print(f2)
```

### W4A28

```python
s = input()
tu_1 = ""
for i in range(len(s)):
    if s[i] == " ":
        break
    tu_1 += s[i] 
print(tu_1)
```

### W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa

```python
s = input()
for i in s:
    if i.islower():
        print(i.upper(), end="")
    elif i.isupper():
        print(i.lower(), end="")
    else:
        print(i, end="")
```

### W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự s

```python
s = input()
hoa = 0
thuong = 0
for i in s:
    if i.isupper():
        hoa += 1
    elif i.islower():
        thuong += 1
print(hoa, "chu hoa", thuong, "chu thuong")
```
