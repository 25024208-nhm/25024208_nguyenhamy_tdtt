### W4A1 [SumOfNumBers]

tính tống các số nguyên từ 1 đến n. (n <= 1000) 

```python
n = int(input())
tong = 0
for i in range(1, n+1):
    tong += i
print(f"Tong tu 1 den {n} la: {tong}")
```

### W4A2 [PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.  
Kiểm tra số này có phải là số nguyên tố hay không?

```python
while True:
    n = int(input())
    if n > 0:
        break
if n <= 1:
    print(f"{n} khong la so nguyen to")
else:
    so_uoc = 0
    for i in range(2, n+1):
        if n % i == 0:
            so_uoc += 1
    if so_uoc == 1:
        print(f"{n} la so nguyen to")
    else:
        print(f"{n} khong la so nguyen to")
```

```python
import math
while True:
    n = int(input())
    if n > 0:
        break
if n <= 1:
    print(f"{n} khong la so nguyen to")
else:
    for i in range(2, math.isqrt(n)+1):
        if n % i == 0:
            print(f"{n} la so nguyen to")
            break
    else:
        print(f"{n} la so nguyen to")
```

### W4A3 [Factorial]
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) với 0 < n < 100.

### W4A4 [CountingDigits]
Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) của số đó. Không dùng kiểu dữ liệu string.


```python
n = int(input())
temp_n = abs(n)
if temp_n == 0:
    count = 1
else:
    count = 0
    while temp_n > 0:
        temp_n //= 10
        count += 1
print(count)
```

### W4A6 [SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn[a, b]. Mỗi dòng là một bộ test (a b).

```python
a, b = map(int, input().split())
def so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
tong = 0
for x in range(a, b+1):
    if so_nguyen_to(x):
        tong += x
print(tong)
```

### W4A7 [LargestPrimeFactor]
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
def uoc_lon_nhat(n):
    max_prime = -1

    while n%2 == 0:
        max_prime = 2
        n //= 2


    for i in range(3, int(n**0.5)+1, 2):
        while n % i == 0:
            max_prime = i
            n //=i

    if n > 2:
        max_prime = n
    
    return max_prime

n = int(input())
print(uoc_lon_nhat(n))
```

### W4A8 [ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu duoc

```python
def reverse(n):
    return int(str(n)[::-1])
def palidrome(n):
    return str(n) == str(n)[::-1]

n = int(input())
steps = 0
while not palidrome(n):
    n = n + reverse(n)
    steps += 1
print(steps, n)
```

### W4A10 [LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.  
Chan: n/2  
Le: 3n+1  
Day Collatz: dua x -> 1


```python
def length(k):
    length = 1
    while k != 1:
        if k % 2 == 0:
            k = k/2
        else:
            k = 3*k+1
        length += 1
    return length

n = int(input())
max_length = 0
best_x = 0
for x in range(1, n+1):
    if length(x) > max_length:
        max_length = length(x)
        best_x = x
print(best_x, max_length)

```

### W4A13 [ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:  
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.  
Ví dụ 220 và 284 là một cặp số thân thiết  
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284  
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220  
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.  
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
def tong_uoc(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

a, b = map(int, input().split())
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("True")
else:
    print("False")
```

### W4A14 [GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n

```python
m, n = map(int, input().split())
ucln = 0
for i in range(1, min(m, n)+1):
    if m % i == 0 and n % i == 0:
        if i > ucln:
            ucln = i
print(ucln)
```
