Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

### W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mang

```python
n = int(input())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
for i in range(n):
    print(f"{matrix[i][i]:4d}", end="")
print()
for i in range(n):
    print(f"{matrix[i][n-1-i]:4d}", end="")
print()
```

### W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
s = input()
words = s.split()
d = {}
for index, word in enumerate(words):
    if word in d:
        d[word] += (index,)
    else:
        d[word] = (index,)
print(d)
```

### W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def unique_array(lst):
    count_zero = lst.count(0)
    non_zero = [x for x in lst if x != 0]
    return [0] * count_zero + non_zero
lst = [0, 2, 3, 2, 4, 0, 0, 0, 9]
print(unique_array(lst))
```

### W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
def sum():
    tong = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            tong += lst[i]
    return tong
lst = list(map(int, input().split()))
print(sum())
```

### W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giam)

```python
lst = list(map(int, input().split()))
Tang = False
Giam = False
for i in range(len(lst)-1):
    if lst[i] < lst[i+1]:
        Tang = True
    elif lst[i] > lst[i+1]:
        Giam = True
if Tang and not Giam:
    print("Tang")
elif Giam and not Tang:
    print("Giam")
else:
    print("Vo danh")
```
