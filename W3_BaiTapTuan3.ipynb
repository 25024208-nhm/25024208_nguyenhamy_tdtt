### W3A11 [Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác.  
Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không.  
Nếu không in ra "Không phải tam giác".  
Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".

```python
a, b, c = map(int, input().split())
if a+b > c and b+c > a and a+c >b:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or c == a:
        print("Tam giac can")
    else:
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")
```

### W3A14 [Giải phương trình bậc nhất]
Nhập vào hai số thực a và b của phương trình ax + b = 0.  
Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng

```python
a, b = map(float, input().split())
if a != 0:
    x = -b / a
    print(f"{x:.2f}")
elif a == b == 0:
    print("Vo so nghiem")
else:
    print("Vo nghiem")
```
