### 1
Yêu cầu. Cước phí Taxi Thông Minh.  
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, kiểu số nguyên, đơn vị VND).  
Các quy tắc cần cài đặt:  
1. Giá mở cửa (0–0.8 km):  
- Xe 4 chỗ: 11,000 VND  
- Xe 7 chỗ: 12,000 VND  
2. Giá theo km sau 0.8 km:  
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km  
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km  
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.  
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + chờ).  
5. Điều kiện đặc biệt:  
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.  
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).   

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    #1. Gia mo cua (0-0.8km)
    if car_type == 4:
        fare = 11000
    elif car_type == 7:
        fare = 12000
    else:
        return 0

    #2. Gia theo km sau 0.8km
    if km > 0.8:
        d = km - 0.8
        if car_type == 4:
            if km <= 20:
                fare += d * 13000
            else:
                fare += (20-0.8)*13000 + (km-20)*10500
        elif car_type == 7:
            if km <= 20:
                fare += d * 14000
            else:
                fare += (20-0.8)*14000 + (km-20)*12000

    #3. Phi cho
    import math
    if wait_minutes > 5:
        extra = wait_minutes - 5
        fare += math.ceil(extra/5)*2000

    #4. Ban dem
    night = start_hour in [22,23,0,1,2,3,4,5]
    if night:
        fare *= 1.2

    #5.DK dac biet
    if km > 30 and not night:
        fare *= 0.9
    if fare < 5000 and km > 5:
        fare = 50000
    
    return int(fare)

car_type, start_hour, km, wait_minutes = map(int, input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

### 14 Viết hàm trả về hệ số nhị thức C(n, k)

```python
def C(n, k):
    if k < 0 or k > n:
        return 0
    if k == 0 or k == n:
        return 1
    if k > n // 2:
        k = n-k
    res = 1
    for i in range(k):
        res = res*(n-i) // (i+1)
    return res
n, k = map(int, input().split())
print(C(n, k))
```

```python
import math
def goc_abc(x1,y1,x2,y2,x3,y3):
    tu = (x1-x2)*(x3-x2) + (y1-y2)*(y3-y2)
    ba = math.sqrt((x1-x2)**2 + (y1-y2)**2)
    bc = math.sqrt((x3-x2)**2 + (y3-y2)**2)
    mau = ba*bc
    return math.degrees(math.acos(tu/mau))
x1,y1,x2,y2,x3,y3 = map(int, input().split())
print(goc_abc(x1,y1,x2,y2,x3,y3))
```

Yêu cầu: Viết hàm point_in_quadrilateral(quad, p) trả về True nếu điểm p nằm bên trong hoặc trên biên của tứ giác quad,
và False nếu ngược lại.  
- Đầu vào:  
- quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm,
nhưng không tự cắt).  
- p: một điểm (x, y).    
- Đầu ra: giá trị boolean.  
- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm
trên biên là ở trong. 


```python
def point_in_quadrilateral(quad, p):
    n = len(quad)
    pos = 0
    neg = 0
    for i in range(n):
        p1 = quad[i]
        p2 = quad[(i+1)%n]
        val = (p2[0] - p1[0])*(p[1]-p1[1]) - (p2[1]-p1[1])*(p[0]-p1[0])
        if val>0:
            pos += 1
        elif val<0:
            neg += 1
        return not (pos>0 and neg>0)
    
quad = eval(input())
p = eval(input())
print(point_in_quadrilateral(quad, p))
```

- Đầu vào: ba điểm x1, y1, x2, y2, x3, y3 biểu diễn tọa độ của ba điểm.  
- Đầu ra: giá trị boolean

Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu chấm động, ngược lại trả về 'neither'.  
- Đầu vào: chuỗi s.  
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.  
- Quy tắc:  
- Cho phép có dấu + hoặc - ở đầu.  
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').  
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').  
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.   

```python
def detect_number_type(s):
    temp = s
    if temp.startswith('+') or temp.startswith('-'):
        temp = temp[1:]
    if temp.isdigit():
        return "int"
    if '.' in temp:
        parts = temp.split('.')
        if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
            return "float"
    return "neither"
s = input()
print(detect_number_type(s))
```

- Yêu cầu thêm:  
- Nếu op là '/' và b == 0, phát sinh ValueError.  
- Nếu op không hợp lệ, phát sinh ValueError.

```python
def compute(a,b,op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b != 0:
            return a/b
    return "ValueError"
a, b = map(int, input().split())
op = input()
print(compute(a,b,op))
```

- Đầu vào: số nguyên dividend, divisor (divisor khác 0).  
- Đầu ra: bộ hai số nguyên (q, r).  
- Ví dụ: divide_with_remainder(5, 3) -> (1, 2).

```python
def divide_with_remainder(dividend, divisor):
    return dividend//divisor, dividend%divisor
dividend, divisor = map(int, input().split())
print(divide_with_remainder(dividend, divisor))
```

### 37 
Viết hàm trả về tam giác vuông cân cao n bằng ký tự ''.  
Dòng thứ i (đánh số từ 1) có i dấu '', các dòng nối với nhau bằng ký tự xuống dòng.  
Ví dụ với n = 4

```python
def tam_giac_vuong_can(n):
    ket_qua = ""
    for  i in range(1,n+1):
        dong = "*" * i
        if i < n:
            ket_qua += dong + "\n"
        else:
            ket_qua += dong
    return ket_qua
n = int(input())
print(tam_giac_vuong_can(n))
```

Ví dụ, n=2, m=3

### 43 
Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể tạo thành tam giác

```python
def S_tam_giac():
    if a+b>c and b+c>a and c+a>b:
        p = (a+b+c)/2
        s = (p*(p-a)*(p-b)*(p-c))**0.5
        return s
    else:
        return "ValueError"
a,b,c = map(float, input().split())
print(S_tam_giac())
```

### 44 
Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1.

```python
def giai_thua(n):
    if n == 0:
        return 1
    else:
        gt = 1
        for i in range(1, n+1):
            gt *= i
        return gt
n = int(input())
print(giai_thua(n))
```

### 45 
Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4). 


```python
def so_luong(n):
    count = 0
    for k in range(1, n):
        if k > (n-1)/2:
            count += 1
    return count
n = int(input())
print(so_luong(n))
```
